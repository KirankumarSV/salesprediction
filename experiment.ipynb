```python
import pandas as pd
from datetime import datetime

def custom_date_parser(x):
    return datetime.strptime(x, '%d-%m-%Y')




salesdata = pd.read_csv("E://Learnbay Project/salesprediction/input/Salesdata.csv", parse_dates= ["Date"], date_parser= custom_date_parser)
salesdata.head()
```

```python
store = pd.read_csv("E://Learnbay Project/salesprediction/input/store.csv")
store.head()
```

```python
sales_data = pd.merge(left= salesdata, right= store, how= "left", on= "Store")
sales_data.head(5)
```

```python
import inflection

snakecase = lambda x : inflection.underscore(x)
new_columns = list(map(snakecase, sales_data.columns))
print("The snake_case columns after renaming will be:  \n \n", new_columns)

#Renaming the columns in snake case
sales_data.columns = new_columns
```

```python
sales_data.to_csv("E://Learnbay Project/salesprediction/input/sales_data.csv", index = False)
```
